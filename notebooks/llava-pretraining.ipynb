```python
import json

def read_json_file(file_path):
    try:
        # Open the file
        with open(file_path, 'r') as file:
            # Parse the JSON content
            data = json.load(file)
        return data
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
    except json.JSONDecodeError as e:
        print(f"Error: Invalid JSON format in '{file_path}'. {str(e)}")
    except Exception as e:
        print(f"An unexpected error occurred: {str(e)}")
    return None

# Example usage
file_path = '/content/drive/MyDrive/ERA v2/multimodal-system/saved_models/llava_image_embeddings.json'
json_data = read_json_file(file_path)

if json_data is not None:
    # Print some information about the loaded data
    print(f"Type of loaded data: {type(json_data)}")

    if isinstance(json_data, dict):
        print(f"Number of keys: {len(json_data)}")
        print("First few keys:")
        for key in list(json_data.keys())[:5]:
            print(f"  {key}")
    elif isinstance(json_data, list):
        print(f"Number of items: {len(json_data)}")
        print("First few items:")
        for item in json_data[:5]:
            print(f"  {item}")

    # If it's a large file, you might want to limit what you print
    print("\nSample of the data:")
    if isinstance(json_data, dict):
        sample_key = next(iter(json_data))
        print(f"{sample_key}: {json_data[sample_key]}...")  # Print first 5 elements
    elif isinstance(json_data, list):
        print(json_data[0] if json_data else "Empty list")
else:
    print("Failed to load JSON data.")
```
